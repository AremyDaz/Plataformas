```python
# Importar pandas
import pandas as pd
import io

# Leer el contenido del archivo
df = pd.read_csv(io.BytesIO(uploaded['file.csv']), index_col=2) # Define una columna como índice
#df = pd.read_csv('file.csv') # No define columna como índice
df.info()
df
```

```python
# Filtros básicos para variables cualitativas
# Filtro por objeto
filtro1=df[df["Entidad_Federativa"] == "Puebla"]
filtro1
```

```python
# Filtro por filas
filtro2 = df.iloc[50:550, : ] #Filas consecutivas
#filtro2=df.iloc[ [1000, 2000, 3000], :] #Filas no consecutivas
filtro2
```

```python
# Filtro por columnas
#filtro3=df.iloc[ : , 2:5] # Columnas consecutivas
filtro3=df.iloc[ : , [1,4,6]] #Columnas no consecutivas
filtro3
```

```python
# Filtro de filas vs columnas
filtro4=df.loc[["Puebla","Durango"], ["Población_ocupada_en_la_industria_manufacturera"]] 
filtro4 #Este filtro requiere columna índice
# NO ME SALIO
```

```python
# Filtro por cabecera
filtro5=df.head()
filtro5
```

```python
# FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS
# Filtro por comparación
filtro6=df[df["Periodo"] >= 2006]
filtro6
```

```python
# FILTROS BASICOS CON CONECTORES LÓGICOS
# Filtro 7
filtro7=df[(df["Periodo"] > 2006) & (df["Sexo"] == "Mujeres")]
filtro7
```

```python
# Filtro OR
filtro8=df[(df["Periodo"] > 2020) | (df["Sexo"] == "Hombre")]
filtro8
```

```python
#Filtro NOT
filtro9=df[~(df["Periodo"] >= 2006) & (df["Sexo"] == "Mujeres")]
filtro9
```

```python
# Convertir archivo filtrado en .csv
filtro9.to_csv("Ejercicio1_Filtros.csv")
```
